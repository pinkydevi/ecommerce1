```python
# Create a function to generate size based on subcategory_id
def generate_size(subcategory_id):
    baby_sizes = ['NB', '0-3M', '3-6M', '6-12M', '12-18M']
    adult_sizes = ['XS', 'S', 'M', 'L', 'XL', 'XXL']
    
    if subcategory_id in [41, 42]:
        return ', '.join(np.random.choice(baby_sizes, np.random.randint(4, 7)))
    elif subcategory_id in [21, 22]:
        return ', '.join(np.random.choice(adult_sizes, np.random.randint(4, 7)))
    else:
        return 'Other'

# Apply the function to each row in 'subcategory_id' column to create 'size' column
df['size'] = df['subcategory_id'].apply(generate_size)
```

```python
# Dictionary mapping subcategory_ids to lists of YouTube video IDs
video_ids_mapping = {
    41: ["cZ56F-BznHc", "tDciKB-84KY","aq4JFSC5cLE", "9pErOxbgAAw","5Hlm-kwNK4E", "VWA7OfmZ0WM"],   # Baby videos
    42: ["A4_Ke3QrPkw", "GGrwdjbbor0","p0Eg833ETJs", "KRA9Tv02X-A","XalumHQlBkM", "DqDFKXAD8iI"],   # Baby videos
    21: ["VYZlIk9tn_c", "j8XI01ePMDE","8gZbHzdjnlI", "MklbIWGOmMs","KoO3iQZ15SU", "bg9iwoQyza8"],   # Adult videos
    22: ["7m16dFI1AF8", "BZDV_esrDgI","knQ4K0Gqm88", "2LCNqRj6tJU","9KhS3VhawPc", "aJXomZeIl_M"]    # Adult videos
}
# Create a function to generate YouTube embed code based on subcategory_id
def generate_youtube_embed(subcategory_id):
    video_ids = video_ids_mapping.get(subcategory_id, [])
    if video_ids:
        youtube_video_id = np.random.choice(video_ids)
        embed_code = youtube_video_id
    else:
        embed_code = "No video available"
    return embed_code

# Apply the function to each row in 'subcategory_id' column to create 'video' column
df['video'] = df['subcategory_id'].apply(generate_youtube_embed)
```

```python
# Create a function to generate random tag based on name
def generate_random_tag(name):
    words = name.split()  # Split the name into words
    tags = [np.random.choice([word.upper(), word.lower(), word.title()]) for word in words]
    return ', '.join(tags)  # Join the tags with a comma

# Apply the function to each row in 'name' column to create 'tags' column
df.loc[:, 'tags'] = df['name'].apply(generate_random_tag)
```
